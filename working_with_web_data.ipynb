## Web Sitemap Scraper

The following Python class, `WebSitemapScraper`, is designed to scrape sitemap data from the given base website. It automatically locates the `robots.txt` file, identifies all the sitemaps, and retrieves the links contained within them. This scraper can be useful for SEO analysis, website structure auditing, or preparing data for further web scraping tasks.


```python
from src.my_packages.WebSitemapScraper import WebSitemapScraper
import pandas as pd
```

## Instantiating the Sitemap Scraper

We will now create an instance of the `WebSitemapScraper` and use it to scrape the sitemap data from the base website. We will then export the aggregated data to a CSV file for further analysis.


```python
# Initialize the scraper with the base URL of the website
scraper = WebSitemapScraper("https://www.datacamp.com")

# Export sitemap data to CSV files
scraper.export_to_csv()

# Combine all sitemap data into a single DataFrame
aggregate_df = pd.concat(scraper.sitemaps_frames.values(), ignore_index=True)
```

## Sitemap Data Overview

The sitemap data has been successfully scraped and aggregated. Below is a preview of the combined sitemap data, which includes the URLs and their respective subdirectories extracted from the sitemap files.


```python
# Display the first few rows of the combined DataFrame
aggregate_df.head()
```

# Part 4: Data Exploration and Visualization

In this section, we will explore the national parks dataset. We will load the dataset, perform basic data exploration, and create visualizations to better understand the data.

## 4.1 Load the Dataset

Let's start by loading the dataset into a Pandas DataFrame:




```python

import pandas as pd

# Load the dataset
df = pd.read_csv("national_parks.csv")
```

To get an initial overview of the dataset, let's display the first few rows:

```python
# Display the first few rows of the dataset
print(df.head())
```

We'll calculate basic statistics for numeric fields:

```python
# Basic statistics of the dataset
print(df.describe())
```

Let's create a visualization to understand the distribution of the 'Best Time to Visit' field:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of the 'Best Time to Visit' field
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Best Time to Visit', order=df['Best Time to Visit'].value_counts().index)
plt.title('Distribution of Best Time to Visit')
plt.xlabel('Count')
plt.ylabel('Best Time to Visit')
plt.show()
```

Let's create a visualization to understand the distribution of national parks by location:

```python
# Location distribution of national parks
plt.figure(figsize=(10, 8))
sns.countplot(data=df, y='Location', order=df['Location'].value_counts().index)
plt.title('Location Distribution of National Parks')
plt.xlabel('Count')
plt.ylabel('Location')
plt.show()
```

This visualization will show the count of national parks in each location. The visualization shows that California has maximum number of National Parks and Alaska is second with Utah and Colorado following.


Let's create a visualization to explore the distribution of top experiences offered by national parks:

```python
# Park Name length distribution
df['Park Name Length'] = df['Park Name'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Park Name Length', bins=20, kde=True)
plt.title('Distribution of Park Name Length')
plt.xlabel('Park Name Length')
plt.ylabel('Frequency')
plt.show()
```
